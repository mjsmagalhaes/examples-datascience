# Avaliation of Forest Fires in Portugal

A brief analysis of Forest Fires data in Portugal.

```python
import pandas as pd
import altair as alt
from dslib.analysis import Analysis
```

```python
raw = pd.read_csv("data/forestfires.csv")
a = Analysis(raw)
```

## Data Description

```python
a.describe()
```

```python
alt.Chart(a.data).mark_bar().encode(
    alt.X('DMC'),
    alt.Y('count()'),
    color='area',
    tooltip='count()'
).interactive()
```

### Charts

```python
chart_1 = alt.Chart(raw).mark_bar().encode(
  x=alt.X('area', bin=alt.BinParams(step=5, extent=[0,50])),
  y='count()',
).interactive()

chart_2 = alt.Chart(raw).mark_bar().encode(
  x=alt.X('area', bin=alt.BinParams(step=50, extent=[50,1100])),
  y='count()',
).interactive()

chart_1 | chart_2
```
